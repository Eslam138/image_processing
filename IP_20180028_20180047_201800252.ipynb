```python
img=cv2.imread('/content/12158644236_bac9086480_b.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
fig, ax = plt.subplots()

ax.imshow(gray)

plt.show()
```

```python
# calculate frequency of pixels in range 0-255
# histg = cv2.calcHist([img],[0],None,[256],[0,256]) 
hist,bins = np.histogram(gray.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
# Create zeros array to store the stretched image
minmax_img = np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image and apply Min-Max formulae
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        minmax_img[i,j] = 255*(gray[i,j]-np.min(gray))/(np.max(gray)-np.min(gray))
 
```

```python
 
hist,bins = np.histogram(minmax_img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(minmax_img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
res = np.hstack((gray,minmax_img)) #stacking images side-by-side


plt.imshow(res)
plt.show()
```

```python

# Example Python Program for contrast stretching


 


def normalize(
    iI      ,
    minI    = 75,
    maxI    = 225,
    minO    = 0,
    maxO    = 255):

    iO      = (iI-minI)*(((maxO-minO)/(maxI-minI))+minO)
    return iO
```

```python
minmax_img2= np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image and apply Min-Max formulae
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        minmax_img2[i,j] = normalize(gray[i,j])
 
```

```python
 
hist,bins = np.histogram(minmax_img2.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(minmax_img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
res = np.hstack((gray,minmax_img,minmax_img2)) #stacking images side-by-side


plt.imshow(res)
plt.show()
```
